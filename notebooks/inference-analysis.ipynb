```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

# First image

```python
idx = 77
pred = '../data/pi/2019_09_04-12_20_24/car_predictions.npy'
image = '../data/pi/2019_09_04-12_20_24/frame0077-12_21_23.jpg'
det_id = 0
```

```python
with open(pred, 'rb') as f:
    for i in range(idx + 1):
        if i != idx:
            np.load(f)
        else:
            arr = np.load(f)
```

```python
arr
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 20)
car = plt.imread(image)
plt.imshow(car)
ax.axis('off')

def what_color(value):
    if value > 0.9:
        return 'red'
    elif value > 0.7:
        return 'orange'
    elif value > 0.5:
        return 'yellow'
    else:
        return 'blue'

for i in range(10):
    if arr[i, 2] < 0.2:
        continue
    top, left, bottom, right = arr[i, 3], arr[i, 2], arr[i, 5], arr[i, 4]
    top, bottom = top * car.shape[0], bottom * car.shape[0]
    left, right = left * car.shape[1], right * car.shape[1]
    color = what_color(arr[i, 1])
    rect = patches.Rectangle((left, bottom), right - left, top - bottom,
                             linewidth=3,edgecolor=color,facecolor='none')
    #ax.annotate(s.loc[int(cl), 0] + ' ' + str(round(conf, 2)), xy=(left, top), color='w', fontsize=15)
    # Add the patch to the Axes
    ax.add_patch(rect)
plt.savefig(image[0:31] + 'det-{}.png'.format(det_id), bbox_inches='tight')
plt.show()
```
